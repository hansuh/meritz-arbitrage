```python
import datetime
import matplotlib.pyplot as plt
```

```python
# Open a file: ./meritz-data/alwayson-log-71456.log
dates=[]
disc_rate_ins=[]
disc_rate_sec=[]
with open("./logs/alwayson-log-71456.log") as file:
    data = file.readlines()
    i=0
    while(i<len(data)-3):
        if data[i].find("=======2022-12-08")>=0 and data[i+2].find("disc_rate_ins")>=0 and data[i+3].find("disc_rate_sec")>=0:
            dates.append(data[i].split("=======")[1])
            disc_rate_ins.append(float(data[i+2].split("=")[1][:-2]))
            disc_rate_sec.append(float(data[i+3].split("=")[1][:-2]))
            i+=3
        i+=1
```

```python
# Convert dates to datetime objects
dates = [datetime.datetime.strptime(date[:-6], '%Y-%m-%d %H:%M:%S.%f') for date in dates]
```

```python
plt.plot(dates,disc_rate_ins)
plt.plot(dates,disc_rate_sec)
plt.legend(['disc_rate_ins','disc_rate_sec'])
plt.xlabel('Time')
plt.show()
```
